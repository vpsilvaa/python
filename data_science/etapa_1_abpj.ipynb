```python
import pandas as pd
import numpy as np

k = 1.380649e-23
T = np.array(range(-10,60,10))
T = T+273.15
Hz = np.array([0.001e3, 1e3, 3e3, 5e3, 10e3, 18e3, 30e3, 100e3, 200e3, 1000e3, 1250e3, 2e9, 3e9, 4e9, 5e9, 10e9, 20e9])
P = np.outer(k*Hz, T)
P_dbm = 10*np.log10(P/1e-3)
df = pd.DataFrame(P_dbm, columns = T, index = Hz)
pd.set_option('display.precision', 3)
df
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x_vals, y_vals = np.meshgrid(T, Hz)
ax = plt.axes(projection = '3d')
ax.plot_surface(x_vals, y_vals, P_dbm, color='red')
ax.set_title('Gráfico dos valores')
plt.show()
```
